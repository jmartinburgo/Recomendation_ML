# Preprocesamiento

```python
df_books_rating = df_books.merge(df_rating,on='ISBN')
df_books_rating.head()
```

```python
df_books_rating.info()
```

```python
#Ranking Libros con mas puntuacion promedio

df_mean_val=df_books_rating.groupby('Book-Title')['Book-Rating'].mean().reset_index().sort_values(by='Book-Rating', ascending=False).rename(columns={'Book-Rating':'Mean'})
df_mean_val
```

```python
df_mean_val.query('`Mean`==10')['Book-Title'][:10].to_list()
```

```python
#Ranking Libros - cantidad de valoraciones
df_cant_val=df_books_rating.groupby('Book-Title')['Book-Rating'].count().reset_index().sort_values(by='Book-Rating',ascending=False).rename(columns={'Book-Rating':'Qtd-Rating'})
df_cant_val.head()
```

```python
#Juntamos los DF proemdios y cantidad de valoraciones
df_feature=df_cant_val.merge(df_mean_val,on='Book-Title')
df_feature.head()
```

```python
#Tomamos solo los que tienen mas de 200 valoraciones
df_feature=df_feature.query('`Qtd-Rating`>200')
df_feature
```

```python
#Aplicamos un Merge entre la tabla original de Books y ratings y la df final. Por defecto el Merge se aplica como Inner, por lo que solo incluyen los valores que comparten
df_final=df_books_rating.merge(df_feature,on='Book-Title')
df_final
```

```python
df_final.isnull().sum()
```

```python
df_final.duplicated().sum()
```

```python
#Creamos una matris que relaciona los libros con lo usuarios a traves del puntaje
df_pivot=df_final.pivot_table(columns='User-ID',index='Book-Title',values='Book-Rating')
df_pivot.head()

```

```python
#Reemplazamos valores Nan por 0
df_pivot.fillna(0,inplace=True)
df_pivot.head()
```

```python
#Transformo el DF en una Matris
from scipy.sparse import csr_matrix

df_sparse = csr_matrix(df_pivot)
df_sparse
```

# Modelo Predictivo

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm='brute')

model.fit(df_sparse)
```

```python
#Tomamos las puntuaciones promedio de la pelicula '1884' y transponemos la lista
pelicula='1984'
mean_movie_selected=df_pivot.loc[pelicula].values.reshape(1,-1)
mean_movie_selected
```

```python
model.kneighbors(mean_movie_selected)
```

```python
num_features = df_sparse.shape[1]
print("Número de características de entrada:", num_features)
```

```python
distances, suggestions=model.kneighbors(mean_movie_selected)

suggestion_final= [j for i in range(len(suggestions)) for j in df_pivot.index[suggestions[i]]]
suggestion_final
```
